In case of training a neural network, how to handle sentences of different length ? -> By Padding

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sentences = [
    'I love my dog',
    'I love my cat',
    'You love my dog!',
    'Do you think my dog is amazing?'
]
```

```python
tokenizer = Tokenizer(num_words = 100, oov_token="<OOV>") #num_words parameter is max words to keep
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index



sequences = tokenizer.texts_to_sequences(sentences)

#test_data = [
#    'i really love my dog',
#    'my dog loves my manatee'
#]

#test_seq = tokenizer.texts_to_sequences(test_data)

padded = pad_sequences(sequences)

#print(test_seq)

print(word_index)

print(sequences)

print(padded)
```

```python
padded = pad_sequences(sequences, padding='post')


print(padded)
```

```python
padded = pad_sequences(sequences, padding='post', maxlen=5)

print(padded)
```

```python
padded = pad_sequences(sequences, padding='post',
                                  truncating='post', maxlen=5)

print(padded)
```
