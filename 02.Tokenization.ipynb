```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
sentences = [
    'I love my dog',
    'I love my cat',
    'You love my dog!',
    'Do you think my dog is amazing?'
]
```

```python
tokenizer = Tokenizer(num_words = 100, oov_token="<OOV>") #num_words parameter is max words to keep
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index



sequences = tokenizer.texts_to_sequences(sentences)

test_data = [
    'i really love my dog',
    'my dog loves my manatee'
]

test_seq = tokenizer.texts_to_sequences(test_data)
print(test_seq)

print(word_index)
```

By using the OOV token property, and setting it as something that you would not expect to see in the corpus, like <OOV> , the tokenizer will create a token for that, and then replace words that it does not recognize with the Out of Vocabulary token instead.
